## cleaning cell for 6milion (need to explore to get value counts)

```python
print('importing...')
crimes = pd.read_csv('crimes.csv')
perps = pd.read_csv('perps.csv')
vics = pd.read_csv('vics.csv')
if len(crimes)+len(perps)+len(vics) > 15000000:
    print('full data acquired')
else:
    print('you imported the samples, doofus')
    
# drop the imported index column
crimes = crimes.drop('Unnamed: 0', axis=1)
perps = perps.drop('Unnamed: 0', axis=1)
vics = vics.drop('Unnamed: 0', axis=1)
    
    
# clean  nulls
print('cleaning nulls...')
perps = perps.applymap(lambda x:None if x == " " else x)
crimes = crimes.applymap(lambda x:None if x == " " else x)
vics = vics.applymap(lambda x:None if x == " " else x)
print('cleaned nulls')

# write out column names
print('decoding columns...')
crimes_cols = [
    'segment_level', 'state', 'agency', 'case_number',
    'date', 'UCR_code', 'attempt_or_complete', 'DAC_1', # drugs, alcohol, or computer equipment used during crime
    'DAC_2', 'DAC_2', 'location_type', 'num_premises_entered',
    'method_entry', 'activity_type_1', 'activity_type_2',
    'activity_type_3', 'weapon_type_1', 'weapon_type_2',
    'weapon_type_3', 'bias_1', 'bias_2', 'bias_3', 'bias_4',
    'bias_5', 'num_common_ORIs'
]
vics_cols = [
    'segment_level', 'state', 'agency', 'case_number',
    'date', 'victim_seq_num', 'ucr_1', 'ucr_2', 'ucr_3',
    'ucr_4', 'ucr_5', 'ucr_6', 'ucr_7', 'ucr_8', 'ucr_9',
    'ucr_10', 'vic_type', 'officer_activity_type',
    'officer_assignment_type', 'officer_other_juris',
    'vic_age', 'vic_sex', 'vic_race', 'vic_ethn',
    'vic_resident_status', 'assault_homic_1', 'assault_homic_2',
    'add_homic_circumstances', 'injury_1', 'injury_2', 'injury_3',
    'injury_4', 'injury_5',
    'offndr_1_rltn', 'vic_offndr_rltn_1', 'offndr_2_rltn', 'vic_offndr_rltn_2',
    'offndr_3_rltn', 'vic_offndr_rltn_3', 'offndr_4_rltn', 'vic_offndr_rltn_4',
    'offndr_5_rltn', 'vic_offndr_rltn_5', 'offndr_6_rltn', 'vic_offndr_rltn_6',
    'offndr_7_rltn', 'vic_offndr_rltn_7', 'offndr_8_rltn', 'vic_offndr_rltn_8',
    'offndr_9_rltn', 'vic_offndr_rltn_9', 'offndr_10_rltn', 'vic_offndr_rltn_10',
    'num_ori_matches',
]
perps_cols = [
    'segment_level', 'state', 'agency', 'case_number', 'date',
    'perp_seq_num', 'perp_age', 'perp_sex', 'perp_race',
    'num_ori_matches', 'perp_ethn'
]
# rename cols
crimes.columns = crimes_cols
perps.columns = perps_cols
vics.columns = vics_cols
print('renamed columns')

# save
crimes.to_csv('crimes_s.csv') # save cleaned samples
perps.to_csv('perps_s.csv')
vics.to_csv('vics_s.csv')
print('cleaned & saved')
```

## Fixing merges: 
I wondered why my df was jumping to 40 million when each df had about 6 million each.
Upon inspection, it looks like case_number is only unique to an agency: they'll use the same case numbers, which makes sense, i suppose, if they just start from 0.
The codebook states that you'll need ORI (agency) as well as case_number to identify a unique record.

```python
crimes_perps = crimes_sub.merge(perps_sub, how='inner', on=['agency','case_number'])
print('merge_1 completed')
```

```python
df = crimes_perps.merge(vics_sub, how='inner', on=['agency','case_number'])
print('merges done')
```
