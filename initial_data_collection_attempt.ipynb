# End initial attempt:

#### This attempt ended in a restructuring of the data-gathering process.
What I really wanted to look at is variables involving the criminal and victim, as well as type of crime, various outcomes for both. Predicting the 'people' side of crime.
The various APIs I'd been investigating were giving me summary statistics, or extremely vague databases with anemic variables. I knew I needed incident-based data.

Fortunately, such a thing exists! The FBI's Uniform Crime Reporting Program organizes and releases _vast_ amounts of data on crime. To my delight, I found that they (somewhat) recently have been releasing reports of their "National Incident-Based Reporting System".

I mucked about to try to API into the NIBRS, to no avail (i don't think they have one...?). Downloading the master file from their website (https://crime-data-explorer.fr.cloud.gov/downloads-and-docs) gave me the data in a massive, unhelpfully encrypted text file. Their site says "Master files for each are fixed‐length, written in an ASCII text format, compressed with WinZip software, and require some programming knowledge to extract data". Cheeky louts.

More googling got me to UMich's sociopolitical research institute (https://www.icpsr.umich.edu/icpsrweb/NACJD/series/128), which has the NIBRS available to download by year. All they need is your full name and email; there was a lot of legalese with scary government words attached, so I complied. There are various regulations on how the data must be used, but this project certainly falls under the category of statistical research (at the most amateur level).

I downloaded the 2016 report (#37065). The .TSVs themselves are rather opaque, and to my great distress, _coded_. There's a codebook attached, which I'm using to translate variable names.
There's separate TSVs for 'offense', 'offender', 'victim', 'arrestee' (and others not of interest). They all seem to contain a case_number column (ex. '2W1I0U7 6288'), so I can work out the variables and join them into a megatable.
I do wonder about incidents where offender and arrestee differ. I could probably estimate the number of such cases by comparing offender_age:arrestee_age or offender_race:arrestee_race matched on case_number.

I'm including the files here in the project folder, edits to be committed as I go. They're quite large - 6.5 million rows in the offender table.
Of note is the number of missing values in many columns. I'll check value_counts for each column and see if they should be dropped.

However, the first task is to rename the columns of interest. The scope of this project will probably be limited by what I'm willing to hard-code/translate as much as what I want to look at.
Renaming columns in Excel isn't terribly arduous, so I'll start with that tomorrow, check value_counts, and drop low-data (high entropy?) columns.
The harder part is translating the entries themselves to data according to the codebook. This can be done by defining a dictionary for each column:
    for row in rows:
        if row[column] in col_dict.keys()
            entry = col_dict[entry]
    so this would change the coding to actual numbers or strings.

I feel like it might. Perhaps that's why they code it in the first place (it certainly isn't for readability).
